```python
label = []

for i, row in tweets.iterrows():
    if '#' in row['text']:
        #print "TEXT:", row['text']
        if(row['label'] == 'racism'):
            label.append([1,0])
        else:
            label.append([0,1])
        #print "------"

# print(label)
f = open("labels.txt", "w")
for i in label:
    f.write(str(i)+'\n')
f.close()
```

```python
# Checks whether a sentence contains a racist word or not

racist_words = ['Hindu','Muslim','Jews','Christian','Muslims','Christian','Hindus','hindus','christians',
                'muslim','muslims','jews']
racist_sentences_flag = []

for i, row in tweets.iterrows():
    if '#' in row['text']:
        #print "TEXT:", row['text']
        words = row['text'].split(' ')
        fl = 0
        for i in words:
            i = i.encode('utf-8')
#             print("i is ",i)
            if i in racist_words:
                print("**found** ",i)
                racist_sentences_flag.append(1)
                fl = 1
                break
    if(fl == 0):
        racist_sentences_flag.append(0)
            
print("**Racist sentences**")            
print(racist_sentences_flag)

```
